```python
new_df=df.copy()
```

```python
new_df['Weight']=np.log(df["Weight"])
```

```python
new_df.head()
```

```python
df2 = pd.get_dummies(new_df,drop_first=True)
```

```python
df2.shape
```

```python
df2.head()
```

### Split the dataframe into x & y

```python
# Separate object for target feature 
y = df2["Weight"]

# Separate object for Predictors 
x= df2.drop("Weight",axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
print('X Shape =',x.shape,'\ny Shape =',y.shape)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape,y_train.shape
```

```python
y_test.shape,y_test.shape
```

### Build the Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()      #calling the linear regression model
lr.fit(x_train,y_train)      #fit is used for start training the model.
y_pred = lr.predict(x_test)
```

### Evaluation Metrics

```python
# Train Score
train_score = lr.score(x_train,y_train)
print('Train Score = ',train_score)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print('R2 Score =',score)
```

### Conclusion

R2 score tells how good my model is performed.Here we can see both train score and test score accuracy are near to each other.So we can say this is a good model.
